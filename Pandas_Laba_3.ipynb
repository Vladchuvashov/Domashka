1 Найти евклидово расстоняние между двумя объектами Series

```python
##Функция linalg.norm() вычисляет норму матрицы.
x = pd.Series([1, 3, 6, 13, 15]) 
y = pd.Series([6, 19, 12, 13, 14]) 
ans = np.linalg.norm(x-y) ##ans = np.sqrt(sum((x-y)**2)) 
print(ans)
```

2 Найти индексы локальных максимумов в объекте Series

```python
sample = pd.Series([1, 5, 7, 11, 8, 4, 12, 5])

# np.diff считает дискретную разницу по след формуле => out[i] = a[i+1] - a[i]
# np.sign возвращает знак в виде цифры  { +x => 1;  -x => -1; 0 => 0}
# в данном случае мы сначала считаем разницу между соседними элементами
# затем переводим в знаки -1, 1, 0
# и затем снова считаем разницу
discrete_difference = np.diff(np.sign(np.diff(sample)))

# В нашем случае локальные максимумы будут там
# где разница между соседними элементами после манипуляций будет равна -2
np.where(discrete_difference == -2)[0] + 1
```

3 Заменить пробелы наименее часто встречающимся символов

```python
s = pd.Series(list('xx y zzz AAAA '))

# value_counts - отдает Serias, содержащую количество уникальных значений
freq = s.value_counts()
# мы берем оттуда самый наименее частый элемент - он последний
least_freq = freq.index[-1]

s.replace(' ', least_freq)
```

4 Создать объект Series, который содержит в индексах даты выходных дней субботы, а в значениях случайные числа от 1 до 10

```python
pd.Series(
    np.random.randint(low=1, high=10, size=[10]),
    pd.date_range('2022-01-01', periods=10, freq='W-SAT'),
)
```

5 Заполнить пропущенные даты, значением выше (заполненной даты)

```python
sample = pd.Series(
    [3, 6, 9, np.nan,12],
    pd.date_range('1998-01-01', periods=5, freq='W-SAT'),
)

# ffill - заполняет предидущими значениями  т.е. [1, 2, nan, 3] => [1, 2, 2, 3]
# bfill - заполняет следующим значениями  т.е.  [1, 2, nan, 3] => [1, 2, 3, 3]

sample.ffill()
```
