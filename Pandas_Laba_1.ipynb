11 Получить не пересекающиеся элементы в двух объектах Series

```python
X = pd.Series([1, 2, 3, 4, 5, 6])
Y = pd.Series([4, 5, 6, 7, 8, 9])
```

12 Возвратить вместе с индексами

```python
pd.concat ([X[~X.isin(Y)],Y[~Y.isin(X)]])
```

13 Получить объединенный Series без повторений

```python
A = pd.Series(np.union1d(X,Y))
A
```

14 Получить пересекающиеся данные

```python
B=pd.Series(np.intersect1d(X,Y))
B
```

15 Отобрать все данные, кроме пересекающихся
