19 Получить год, месяц, день, день недели, номер дня в году от объекта Series (string)

26 Отобрать элементы объекта Series, которые содержат не менее двух гласных

```python
from collections import Counter
 
s = pd.Series(['Vladislav', 'Baby', 'dont', 'hurt', 'me', 'Моя', 'Любимая', 'Песня'])
mask = s.map(lambda x: sum([Counter(x.lower()).get(i, 0) for i in list('aieouyаеёиоуыэюя')]) >= 2)
ans = s[mask]
print(ans)
```

27 Отобрать e-маилы из объекта Series (можно юзать регулярки)

```python
import re
 
emails = pd.Series(['test text @test.com', 'test@mail.ru', 'test.2ru', 'test@pp'])
pattern = '[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'
mask = emails.map(lambda x: bool(re.match(pattern, x)))
ans = emails[mask]
print(ans)
```

28 Получить среднее значение каждого уникального объекта Series s1 через "маску" другого объекта Series s2

```python
n = 10
s1 = pd.Series(np.random.choice(['dog', 'cat', 'horse', 'bird'], n))
s2 = pd.Series(np.linspace(1,n,n))
ans = s2.groupby(s1).mean()
print(ans)
```
